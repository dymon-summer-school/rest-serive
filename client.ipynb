```python
import geopandas as gpd
import folium
import requests
```

```python
# fetch bundesland
bundesland: dict = requests.get('http://dss.trafficon.eu:4000/bundesland/Wien').json()
#bundesland: dict = requests.get('http://0.0.0.0:4000/bundesland/Salzburg').json()

# create map
m = folium.Map(location=[47.5, 13.4], zoom_start=7, tiles="cartodbpositron")
# add bundesland
geo_j = folium.GeoJson(data=bundesland, style_function=lambda x: {"fillColor": "blue", "color": "#234345"})
geo_j.add_to(m)

# draw map
m
```
